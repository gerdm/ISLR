```python
import numpy as np 
from matplotlib import style, pyplot as plt
from numpy.random import normal, binomial, poisson
```

# KNN

The test error rate associated with a set of test obeservations of the form $(x_0, y_0)$: $\text{Ave}(I(y_0 \neq \hat{y}_0))$ is minimized, on average, by a classifier that assigns each observation to the most likely class, given its predictor values. Then, we should assign a test observation with predictor vector $x_0$ to the class $j$ for which

$$
   \mathbb{P}(Y=j|X=x_0)
$$

is largest

```python
nsamples = 400
X1 = normal(3, 2, nsamples)
X2 = normal(-2, 2, nsamples)

def KNN(x1,x2, X1, X2, K=1):
    distances = np.sqrt((X1 - x1) ** 2 + (X2 -  x2) ** 2)
    K_nearest = distances.argsort()[:K]
    sample = np.array([class_sample[k] for k in K_nearest])
    kclass = np.mean(sample)
        
    if kclass < 0.5:
        return 0
    else:
        return 1
    
    
def classifier(x1, x2):
    clf = np.sin(x1) + np.cos(x2) > 0
    #clf = binomial(1, 0.7, 1)
    return clf 

class_sample = []

fig = plt.figure(figsize=(10,7))
for points in zip(X1, X2):
    
    if classifier(points[0], points[1]):
        class_sample.append(1)
        plt.scatter(*points, c="orange")
        
    else:
        class_sample.append(0)
        plt.scatter(*points, c = "ccc")


xrange = np.linspace(min(X1), max(X1), 100)
yrange = np.linspace(min(X2), max(X2), 100)

for x in xrange:
    for y in yrange:
        dotknn = KNN(x, y, X1, X2, K=15)
        color = "orange" if dotknn == 1 else "ccc"
        plt.scatter(x, y, s = 2, edgecolors=color)
        


plt.xlim(min(X1) -0.1, max(X1) + 0.1)
plt.ylim(min(X2) - 0.1, max(X2) + 0.1)
plt.xlabel(r"$X_1$")
plt.ylabel(r"$X_2$")
plt.show()
```
