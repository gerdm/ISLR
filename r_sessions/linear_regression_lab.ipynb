## Simple Linear Regression

```r
names(Boston)
```

**Predicting median value of home given the percent of lower status in the population**

```r
plot(medv ~ lstat, Boston)
```

```r
fit1 = lm(medv ~ lstat, Boston)
summary(fit1)
```

```r
plot(medv ~ lstat, Boston)
abline(fit1, col="red")
```

```r
confint(fit1)
```

```r
predict(fit1, data.frame(lstat=c(5,10,15)), interval="confidence")
```

## Multiple Linear Regression

```r
fit2 = lm(medv ~ lstat + age, data=Boston)
summary(fit2)
```

**Predicting medv by all other variables in the dataset**

```r
fit3 = lm(medv ~ ., Boston)
summary(fit3)
```

```r
par(mfrow=c(2,2))
plot(fit3)
```

**Taking only the significant Factors**

```r
# Same response, same predictors except 'age' and 'indus'
fit4 = update(fit3, ~ . - age - indus)
summary(fit4)
```

## Nonlinear Terms and Interactions

```r
# Interaction
fit5 = lm(medv~lstat*age, Boston)
summary(fit5)
```

```r
# Nonlinear Term
fit6 = lm(medv~lstat + I(lstat^2), Boston)
summary(fit6)
```

```r
par(mfrow=c(1,1))
with(Boston, plot(medv~lstat))
points(Boston$lstat, fitted(fit6), col="red", pch=20)
```

**Fitting a 4th degree polynomia**

```r
#attach(Boston)
fit7 = lm(medv~poly(lstat, 4))
plot(medv ~ lstat)b
points(lstat, fitted(fit6), col="red", pch=20)
points(lstat, fitted(fit7), col="blue", pch=20)
```
